## Let's get started!

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('turnstile_180901.txt')
print(df.shape)
df.head()
```

## Rename all the columns to lower case

```python
replacement_columns = []
for item in df.columns:
    replacement_columns.append(item.lower())
df.columns = replacement_columns
df.columns
    
```

## Change the Index to be the Line Names

```python
df.set_index('linename')
```

## Remove the index

```python
df.reset_index()
```

## Create another column 'Num_Lines' that is a count of how many lines pass through a station. Then sort your DataFrame by this column in descending order
*Hint: According to the [data dictionary](http://web.mta.info/developers/resources/nyct/turnstile/ts_Field_Description.txt), LINENAME represents all train lines that can be boarded at a given station. Normally lines are represented by one character. For example, LINENAME 456NQR represents trains 4, 5, 6, N, Q, and R.*

```python
def linelen (string):
    output = len(string)
    return output

df['Num_Lines']=df.linename.map(linelen)
df.sort_values(by=['Num_Lines'])
df
```

## Write a function to clean a column name

```python
def clean(col_name):
    cleaned = col_name.capitalize().strip()
    return cleaned
```

```python
# This is a list comprehension. It applies your clean function to every item in the list.
# We then reassign that to df.columns
# You shouldn't have to change anything here.
# Your function above should work appropriately here.
df.columns = [clean(col) for col in df.columns] 
df.columns
```

```python
# Checking the output, we can see the results.
df.columns
```

## Group the Data by Day of Week and Plot the Sum of The Numeric Columns

```python
df.Date = pd.to_datetime(df.Date)
df.Date.dt.day_name()
df['Day']=df.Date.dt.day_name()
day_freq = df.Day.value_counts()
day_freq.plot.bar()
```

## Group the Data by Weekend/Weekday and Plot the Sum of the Numeric Columns

```python
df.loc[df["Day"] == 'Monday', "DayType"] = "Weekday"
df.loc[df["Day"] == 'Tuesday', "DayType"] = "Weekday"
df.loc[df["Day"] == 'Wednesday', "DayType"] = "Weekday"
df.loc[df["Day"] == 'Thursday', "DayType"] = "Weekday"
df.loc[df["Day"] == 'Friday', "DayType"] = "Weekday"
df.loc[df["Day"] == 'Saturday', "DayType"] = "Weekend"
df.loc[df["Day"] == 'Sunday', "DayType"] = "Weekend"
weekend_freq = df.DayType.value_counts()
weekend_freq.plot.bar()
```

## Analysis Question: 

What is misleading about the day of week and weekend/weekday charts you just plotted?

```python
# The weekend / weekday chart is misleading because there are many more days counted in the weekday bar vs the weekend bar
# The day chart is misleading because the count of entries / exits per line varies by a factor of almost 10,000 and it doesn't capture that volume variability
```
